```python
from strictjson import *
import pandas as pd
from pprint import pprint
```

```python
df = pd.read_csv('../data/india_top_50_cities.csv')
df.head()
```

```python
# df.loc[df['City'] == 'before', 'City'] = 'after'
```

```python
categories = [
'Historical Sites',
'Mueseums',
'Art Gallery',
'Piligrimage',
'Beaches',
'Hiking',
'Greenery',
'Mountains',
'Shopping',
'Nightlife',
'Urban',
'Desert',
'Adventure',
'Cultural Experience',
'Waterfalls',
'Caves',
'Wildlife Experience',
'Skiing',
'Scuba Diving',
'Water Activites']
```

```python
df = pd.concat([df, pd.DataFrame({category: [] for category in categories})], axis=1)
df.head()
```

```python
def get_categories_presence(place_name):
    res = strict_json(
    system_prompt="You are a classifier",
    user_prompt=place_name,
    output_format={
        "Historical Sites":"Wether or not this place has Historical Sites",
        "Mueseums":"Wether or not this place has Mueseums",
        "Art Gallery":"Wether or not this place has Art Galleries",
        "Piligrimage":"Wether or not this place has Piligrimages",
        "Beaches":"Wether or not this place has Beaches",
        "Hiking":"Wether or not this place has Hiking",
        "Greenery":"Wether or not this place has Greenery",
        "Mountains":"Wether or not this place has Mountains",
        "Shopping":"Wether or not this place has Shopping",
        "Nightlife":"Wether or not this place has Nightlife",
        "Urban":"Wether or not this place is Urban",
        "Desert":"Wether or not this place has Desert",
        "Adventure":"Wether or not this place can provide an Adventurous Experience",
        "Cultural Experience":"Wether or not this place can provide a Cultural Experience",
        "Waterfalls":"Wether or not this place has Waterfalls",
        "Caves":"Wether or not this place has Caves",
        "Wildlife Experience":"Wether or not this place can provide a Wildlife Experience",
        "Skiing":"Wether or not this place has Skiing",
        "Scuba Diving":"Wether or not this place has Scuba Diving Facilities",
        "Water Activites":"Wether or not this place has Water Activites",
    },
)
    return res
```

```python
print(get_categories_presence('Tirupati'))
```

```python
for index, row in df.iterrows():
    try:
        city = row['City']
        categories_presence = get_categories_presence(city)
        for category, presence in categories_presence.items():
            df.loc[index, category] = 1 if presence == 'Yes' else 0
    except:
        continue
    # city = row['City']
    # categories_presence = get_categories_presence(city)
    # for category, presence in categories_presence.items():
    #     df.loc[index, category] = 1 if presence == 'Yes' else 0
    
```
